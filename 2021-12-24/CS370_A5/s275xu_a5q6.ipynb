# A5-Q6: Audio Enhancing

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile
from IPython.display import Audio
```

```python
# Load the audio recording
Omega, f = scipy.io.wavfile.read('recording.wav')
Audio(f, rate=Omega)
```

```python
# Some useful values
N = len(f)    # total number of samples
L = N / Omega # length of sound clip (in seconds)
t = np.arange(0,N-1) * L/N  # array of time stamps for samples
```

```python
# Corresponding array of sampled frequencies
omega = np.fft.fftshift(np.arange(-N/2, N/2)) / L
```

```python
plt.plot(f)
plt.title('Sound')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude');
```

## (a)

```python
# === YOUR CODE HERE ===
# DFT of f
F = np.fft.fft(f)
plt.plot(omega, abs(F))
plt.title('Frequency domain of F')
plt.xlabel('Hz')
plt.ylabel('Modulus')
```

## (b)

```python
# === YOUR CODE HERE ===
# threshold frequency
threshold = 2200
mask = abs(omega) < threshold
Flow = F * mask
plt.plot(omega, abs(Flow))
plt.title('Adjusted')
plt.xlabel('Hz')
plt.ylabel('Modulus')
```

## (c)

```python
# === YOUR CODE HERE ===
flow = np.fft.ifft(Flow)
plt.plot(np.real(flow))
plt.title('Filtered')
plt.xlabel('s')
plt.ylabel('A')
```

```python
# Play the audio clip
# === YOUR CODE HERE ===
Audio(np.real(flow), rate=Omega)
```
